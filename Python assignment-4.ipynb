```python
'''
1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns
the list of words that are longer than n.
'''
def filter_long_words(list1,n):
    list2=[]
    for i in list1:
        if len(i)>n:
            list2.append(i)
    else:
        print(list2)
```

```python
filter_long_words(['aaaaq','nnjdkjfk','kd','jdkdjk','34'],2)
```

```python
'''
2.1 Write a Python program using function concept that maps list of words into a list of integers
representing the lengths of the corresponding words.
'''
def lengthofwords(num):
    return len(num)
list1=['aaaaq','nnjdkjfk','kd','jdkdjk','34']
print(list(map(lengthofwords,list1)))
```

```python
'''
Write a Python function which takes a character (i.e. a string of length 1) and returns True if
it is a vowel, False otherwise.
'''
vowels_list=['a','e','i','o','u']
def vowelornot(ch):
    if ch in vowels_list:
        return True
    else:
        return False
print(vowelornot('z'))
print(vowelornot('e'))
```

```python
'''
parent class and sub class
'''
class parent:
    def __init__(self,s1,s2,s3):
        self.s1=s1
        self.s2=s2
        self.s3=s3
class child(parent):    
    def cal(self):
        s=(self.s1+self.s2+self.s3)/2
        area=(s*(s-self.s1)*(s-self.s2)*(s-self.s3))**0.5
        print("Area of a triangle is : ",area)
a=eval(input("Enter value of side 1 : "))
b=eval(input("Enter value of side 2 : "))
c=eval(input("Enter value of side 3 : "))
objchild=child(a,b,c)
objchild.cal()
```
